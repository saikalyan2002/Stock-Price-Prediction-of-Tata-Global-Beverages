# Stock Price Prediction Using Machine Learning Algorithms
Predicting the future direction of stock prices is a difficult task. The prices depends on many factors like the market's opinion about a stock, latest announcements, etc. We will use machine learning to use this information and try to predict the stock prices.<br>
The methods used are:-
1. Moving Average
2. Linear Regression
3. K-Nearest Neighbors
4. Auto Arima
5. Long short term memory

### Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 20,10
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
df = pd.read_csv('NSE-TATAGLOBAL11.csv')
df.head()
```

Closing price is the weighted average price of the last 30 mins of trading, whereas 'Last' is the last traded price of the stock on that day.<br>
1. Total Trade Quantity is the number of shares bought or sold in the day.
2. Turnover is the turnover of the particular company.

```python
df['Date'] = pd.to_datetime(df.Date,format='%Y-%m-%d')
```

```python
df.index = df['Date']
```

```python
plt.figure(figsize=(16,8))
plt.plot(df['Close'],label='Closing prices')
```

# Moving Average

```python
data = df.sort_index(ascending=True,axis=0)
```

```python
data.head()
```

```python
new_data = pd.DataFrame(index=range(0,len(df)),columns=['Date','Close'])
```

```python
for i in range(0,len(data)):
    new_data['Date'][i] = data['Date'][i]
    new_data['Close'][i] = data['Close'][i]
```

```python
train = new_data[:987].copy()
valid = new_data[987:].copy()
```

```python
train.shape
```

```python
valid.shape
```

```python
preds = []
for i in range(0,valid.shape[0]):
    a = train['Close'][len(train)-248+i:].sum() + sum(preds)
    b = a/248
    preds.append(b)
```

```python
rms = np.sqrt(np.mean(np.power((np.array(valid['Close'])-preds),2)))
print('\n RMSE value on validation set:')
print(rms)
```

```python
valid['Predictions'] = 0
valid['Predictions'] = preds
plt.plot(train['Close'],label='Train')
plt.plot(valid[['Close','Predictions']],label='Valid')
```

# Linear Regression

```python
df['Date'] = pd.to_datetime(df.Date,format='%Y-%m-%d')
df.index = df['Date']
```

```python
data = df.sort_index(ascending=True, axis=0)
```

```python
new_data = pd.DataFrame(index=range(0,len(df)),columns=['Date', 'Close'])
```

```python
for i in range(0,len(data)):
    new_data['Date'][i] = data['Date'][i]
    new_data['Close'][i] = data['Close'][i]
```

```python
from fastai.tabular.core import  add_datepart
add_datepart(new_data, 'Date')
new_data.drop('Elapsed', axis=1, inplace=True) 
```

```python
new_data.head()
```

```python
new_data['mon_fri'] = 0
for i in range(0,len(new_data)):
    if (new_data.iloc[i,5]==0 or new_data.iloc[i,5]==4):
        new_data.iloc[i,-1] = 1
    else:
        new_data.iloc[i,-1] = 0
```

```python
train = new_data[:987].copy()
valid = new_data[987:].copy()
```

```python
x_train = train.drop('Close', axis=1)
y_train = train['Close']
x_valid = valid.drop('Close', axis=1)
y_valid = valid['Close']
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
preds = model.predict(x_valid)
rms = np.sqrt(np.mean(np.power((np.array(y_valid)-np.array(preds)),2)))
rms
```

```python
valid['Predictions'] = 0
valid['Predictions'] = preds
```

```python
valid.index = new_data[987:].index
train.index = new_data[:987].index
```

```python
plt.plot(train['Close'])
plt.plot(valid[['Close', 'Predictions']])
```

# K-Nearest Neighbors

```python
from sklearn import neighbors
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
```

```python
train = new_data[:987].copy()
valid = new_data[987:].copy()
```

```python
x_train = train.drop('Close', axis=1)
y_train = train['Close']
x_valid = valid.drop('Close', axis=1)
y_valid = valid['Close']
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_train = pd.DataFrame(x_train_scaled)
x_valid_scaled = scaler.fit_transform(x_valid)
x_valid = pd.DataFrame(x_valid_scaled)
```

```python
params = {'n_neighbors':[2,3,4,5,6,7,8,9]}
knn = neighbors.KNeighborsRegressor()
model = GridSearchCV(knn, params, cv=5)
```

```python
model.fit(x_train,y_train)
preds = model.predict(x_valid)
```

```python
rms=np.sqrt(np.mean(np.power((np.array(y_valid)-np.array(preds)),2)))
rms
```

```python
valid['Predictions'] = 0
valid['Predictions'] = preds
```

```python
plt.plot(valid[['Close', 'Predictions']])
plt.plot(train['Close'])
```

# ARIMA


```python
from pmdarima import auto_arima
```

```python
data = df.sort_index(ascending=True, axis=0)
```

```python
train = data[:987].copy()
valid = data[987:].copy()
```

```python
training = train['Close']
validation = valid['Close']
```

```python
model = auto_arima(training, start_p=1, start_q=1,max_p=3, max_q=3, m=12,start_P=0, seasonal=True,d=1, D=1, trace=True,error_action='ignore',suppress_warnings=True)
model.fit(training)
```

```python
forecast = model.predict(n_periods=248)
forecast = pd.DataFrame(forecast,index = valid.index,columns=['Prediction'])
```

```python
rms=np.sqrt(np.mean(np.power((np.array(valid['Close'])-np.array(forecast['Prediction'])),2)))
rms
```

```python
plt.plot(train['Close'])
plt.plot(valid['Close'])
plt.plot(forecast['Prediction'])
```

# LSTM

```python
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM
```

```python
data = df.sort_index(ascending=True, axis=0)
new_data = pd.DataFrame(index=range(0,len(df)),columns=['Date', 'Close'])
for i in range(0,len(data)):
    new_data['Date'][i] = data['Date'][i]
    new_data['Close'][i] = data['Close'][i]
```

```python
new_data.index = new_data.Date
new_data.drop('Date', axis=1, inplace=True)
```

```python
new_data.head()
```

```python
dataset = new_data.values
```

```python
dataset
```

```python
train = dataset[0:987,:].copy()
valid = dataset[987:,:].copy()
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(dataset)
```

```python
x_train, y_train = [], []
for i in range(100,len(train)):
    x_train.append(scaled_data[i-100:i,0])
    y_train.append(scaled_data[i,0])
x_train, y_train = np.array(x_train), np.array(y_train)
x_train = np.reshape(x_train, (x_train.shape[0],x_train.shape[1],1))
```

```python
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(x_train.shape[1],1)))
model.add(LSTM(units=25))
model.add(Dense(1))
```

```python
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(x_train, y_train, epochs=10, batch_size=10, verbose=2)
```

```python
inputs = new_data[len(new_data) - len(valid) - 100:].values
```

```python
inputs.shape
```

```python
inputs = inputs.reshape(-1,1)
print(inputs.shape)
inputs  = scaler.transform(inputs)
```

```python
X_test = []
for i in range(100,inputs.shape[0]):
    X_test.append(inputs[i-100:i,0])
X_test = np.array(X_test)
```

```python
X_test = np.reshape(X_test, (X_test.shape[0],X_test.shape[1],1))
print(X_test.shape)
closing_price = model.predict(X_test)
print(closing_price.shape)
closing_price = scaler.inverse_transform(closing_price)
```

```python
rms=np.sqrt(np.mean(np.power((valid-closing_price),2)))
rms
```

```python
train = new_data[:987].copy()
valid = new_data[987:].copy()
```

```python
valid['Predictions'] = closing_price
plt.plot(train['Close'])
plt.plot(valid[['Close','Predictions']])
```
